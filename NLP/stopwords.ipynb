```python
paragraph = '''
Gojo Satoru, from *Jujutsu Kaisen*, embodies qualities that resonate beyond the realm of anime. His character—marked by unparalleled strength, charisma, and a sense of justice—serves as a metaphor for confronting challenges with confidence and resilience. In real life, Gojo's unyielding determination and ability to inspire others highlight the importance of maintaining a strong sense of purpose and leadership in the face of adversity. His role encourages individuals to embrace their unique strengths and to confront obstacles with a mindset geared toward positive change and personal growth. By reflecting on Gojo's character, people can find motivation to persevere through their own struggles and to lead with integrity and conviction.
'''
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords,wordnet
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
stopwords.words('english')
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmetizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmetizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```
