```python
sentences = "Tony Stark is also known as the invinvible iron man. He is the creator of Jarvis AI. He is also the owner of the stark industries"
```

```python
import nltk
words = nltk.word_tokenize(sentences)
```

```python
tagged_words = nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tagged_words).draw()
```
